### Functions

````python
# Image preprocessing
def standardize(input_array:np.array):
  input_array = np.nan_to_num(input_array, copy=False,)
  return (input_array - np.mean(input_array)) / np.std(input_array)

def normalize(input_array:np.array): # Min-Max Normalization
  input_array = np.nan_to_num(input_array, copy=False,)
  return (input_array - np.min(input_array)) / (np.max(input_array) - np.min(input_array)) * 255.

# Image file loading
def read_NIBImage (input_path:str):
  image = nib.load(input_path).get_fdata()
  # image = standardize(image)
  image = normalize(image)
  return image
````

````python
# Define transforms
train_transforms = Compose([
    SignalFillEmpty(),
    ScaleIntensity(),
    EnsureChannelFirst(),
    Resize(win_size),
    RandRotate90(),
])

val_transforms = Compose([
    SignalFillEmpty(),
    ScaleIntensity(),
    EnsureChannelFirst(),
    Resize(win_size)
])

# Define nifti dataset, data loader
_str_ImagePaths = df_train['Images_Files'].apply(lambda x:os.path.join(str_TrainDatasetPath,x))
# print(myFun.read_NIBImage(_str_ImagePaths[0]))
check_ds = ImageDataset(image_files= _str_ImagePaths,
                        labels=Onehot_trainLabel,
                        reader= "NibabelReader",
                        transform=train_transforms)
check_loader = DataLoader(check_ds,
                          batch_size=5,
                          num_workers=5,
                          pin_memory=pin_memory)

im, label = monai.utils.misc.first(check_loader)
print(type(im), im.shape, label, label.shape)
# print(im)

# create a training data loader
_str_ImagePaths = df_train['Images_Files'].apply(lambda x:os.path.join(str_TrainDatasetPath,x))
train_ds = ImageDataset(image_files=_str_ImagePaths,
                        labels=Onehot_trainLabel,
                        reader= "NibabelReader",
                        transform=train_transforms)
train_loader = DataLoader(train_ds,
                          batch_size= int_BatchSize,
                          shuffle=True,
                          num_workers=int_BatchSize,
                          pin_memory=pin_memory)

# create a validation data loader
_str_ImagePaths = df_valid['Images_Files'].apply(lambda x:os.path.join(str_TrainDatasetPath,x))
val_ds = ImageDataset(image_files=_str_ImagePaths,
                      labels=Onehot_validLabel,
                      reader= "NibabelReader",
                      transform=val_transforms)
val_loader = DataLoader(val_ds,
                        batch_size=int_BatchSize,
                        num_workers=int_BatchSize,
                        pin_memory=pin_memory)

# create a testing data loader
_str_ImagePaths = df_test['Images_Files'].apply(lambda x:os.path.join(str_TestDatasetPath,x))
test_ds = ImageDataset(image_files=_str_ImagePaths,
                      labels=Onehot_testLabel,
                      reader= "NibabelReader",
                      transform=val_transforms)
test_loader = DataLoader(test_ds,
                        batch_size=int_BatchSize,
                        num_workers=int_BatchSize,
                        pin_memory=pin_memory)
````

````python
# Create DenseNet121, CrossEntropyLoss and Adam optimizer
NetArch = monai.networks.nets.DenseNet121(
    spatial_dims=3,
    in_channels=1,
    out_channels=len(dict_Label2Num.keys()),
)
model=NetArch.to(device)

# loss_function = torch.nn.CrossEntropyLoss()
loss_function = torch.nn.BCEWithLogitsLoss()  # also works with this data

optimizer = torch.optim.Adam(model.parameters(), float_LearningRate)
````

## Model Training

````python
# start a typical PyTorch training
val_interval = 1
best_metric = -1
best_metric_epoch = -1
epoch_loss_values = []
metric_values = []
# scaler = torch.cuda.amp.GradScaler()
writer = SummaryWriter()

for epoch in range(int_Epochs):
    print("-" * 10)
    print(f"epoch {epoch + 1}/{int_Epochs}")
    model.train()
    epoch_loss = 0
    step = 0

    for batch_data in train_loader:
        step += 1
        inputs, labels = batch_data[0].to(device), batch_data[1].to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = loss_function(outputs, labels)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()
        epoch_len = len(train_ds) // train_loader.batch_size
        print(f"{step}/{epoch_len}, train_loss: {loss.item():.4f}")
        writer.add_scalar("train_loss", loss.item(), epoch_len * epoch + step)

    epoch_loss /= step
    epoch_loss_values.append(epoch_loss)
    print(f"epoch {epoch + 1} average loss: {epoch_loss:.4f}")

    if (epoch + 1) % val_interval == 0:
        model.eval()

        num_correct = 0.0
        metric_count = 0
        # for val_data in val_loader:
        for val_data in val_loader:
            val_images, val_labels = val_data[0].to(device), val_data[1].to(device)
            with torch.no_grad():
                val_outputs = model(val_images)
                value = torch.eq(val_outputs.argmax(dim=1), val_labels.argmax(dim=1))
                metric_count += len(value)
                num_correct += value.sum().item()

        metric = num_correct / metric_count
        metric_values.append(metric)

        if metric >= best_metric:
            best_metric = metric
            best_metric_epoch = epoch + 1
            torch.save(model.state_dict(), "best_metric_model_classification3d_array(AD).pth")
            print("saved new best metric model")

        print(f"Current epoch: {epoch+1} current accuracy: {metric:.4f} ")
        print(f"Best accuracy: {best_metric:.4f} at epoch {best_metric_epoch}")
        writer.add_scalar("val_accuracy", metric, epoch + 1)

print(f"Training completed, best_metric: {best_metric:.4f} at epoch: {best_metric_epoch}")
writer.close()
````

````python
plt.plot(epoch_loss_values, label="training loss")
val_epochs = np.linspace(1, int_Epochs, np.floor(int_Epochs / val_interval).astype(np.int32))
plt.plot(val_epochs, metric_values, label="validation acc")
plt.legend()
plt.xlabel("Epoch")
plt.ylabel("Value")
plt.show()
````

# Interpretability

Use GradCAM and occlusion sensitivity for network interpretability.
The occlusion sensitivity returns two images: the sensitivity image and the most probable class.
* Sensitivity ima
ge -- how the probability of an inferred class changes as the corresponding part of the image is occluded.
    * Big decreases in the probability imply that that region was important in inferring the given class
    * The output is the same as the input, with an extra dimension of size N appended. Here, N is the number of inferred classes. To then see the sensitivity image of the class we're interested (maybe the true class, maybe the predcited class, maybe anything else), we simply do ``im[...,i]``.
* Most probable class -- if that part of the image is covered up, does the predicted class change, and if so, to what? This feature is not used in this notebook.

````python
# cam = monai.visualize.CAM(nn_module=model, target_layers="class_layers.relu", fc_layers="class_layers.out")
cam = monai.visualize.GradCAM(nn_module=model, target_layers="class_layers.relu")
# cam = monai.visualize.GradCAMpp(nn_module=model, target_layers="class_layers.relu")
print(
    "original feature shape",
    cam.feature_map_size([1, 1] + list(win_size), device),
)
print("upsampled feature shape", [1, 1] + list(win_size))

occ_sens = monai.visualize.OcclusionSensitivity(nn_module=model, mask_size=12, n_batch=1)

# For occlusion sensitivity, inference must be run many times. Hence, we can use a
# bounding box to limit it to a 2D plane of interest (z=the_slice) where each of
# the arguments are the min and max for each of the spatial dimensions (in this case HWD).
the_slice = test_ds[0][0].shape[-1] // 2
# the_slice = 60
occ_sens_b_box = [-1, -1, -1, -1, the_slice - 1, the_slice]
````

## Patients

````python
train_transforms.set_random_state(5)
n_examples = 10
subplot_shape = [3, n_examples]
plt.figure()
fig, axes = plt.subplots(*subplot_shape, figsize=(50, 15), facecolor="white")
items = np.random.choice(len(test_ds), size=len(test_ds), replace=False)

example = 0
for item in items:
    data = test_ds[item]  # this fetches training data with random augmentations
    image, label = data[0].to(device).unsqueeze(0), data[1][1]
    y_pred = model(image)
    pred_label = y_pred.argmax(1).item()
    # Only display tumours images
    if label == dict_Label2Num['Normal'] or label != pred_label:
        continue

    img = image.detach().cpu().numpy()[..., the_slice]
    # print(item,y_pred,label,data[1])
    name = "actual: "
    name += f"{dict_Num2Label[1]}" if label == 1 else f"{dict_Num2Label[0]}"
    name += "\npred: "
    name += f"{dict_Num2Label[1]}" if pred_label == 1 else f"{dict_Num2Label[0]}"
    name += f"\n{dict_Num2Label[1]}: {y_pred[0,1]:.3}"
    name += f"\n{dict_Num2Label[0]}: {y_pred[0,0]:.3}"

    # run CAM
    cam_result = cam(x=image, class_idx=None)
    cam_result = cam_result[..., the_slice]

    # run occlusion
    occ_result, _ = occ_sens(x=image, b_box=occ_sens_b_box)
    occ_result = occ_result[0, pred_label][None, None, ..., -1]

    for row, (im, title) in enumerate(
        zip(
            [img, cam_result, occ_result],
            [name, "CAM", "Occ. sens."],
        )
    ):
        cmap = "gray" if row == 0 else "jet"
        ax = axes[row, example]
        if isinstance(im, torch.Tensor):
            im = im.cpu().detach()
        im_show = ax.imshow(im[0][0], cmap=cmap)

        ax.set_title(title, fontsize=25)
        ax.axis("off")
        fig.colorbar(im_show, ax=ax)

    example += 1
    if example == n_examples:
        break
plt.show()
````
